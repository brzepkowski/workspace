

# Introduction
Support vector machines so called as SVM is a supervised learning algorithm which can be used for classification and regression problems. It is used for smaller dataset as it takes too long to process. SVM is based on the idea of finding a hyperplane that best separates the features into different domains.


# Terminology




![tekst alternatywny](https://drive.google.com/uc?export=view&id=10AjhoZlB5CStmf9m34Nvmcx3oPl2ifu_)

The points closest to the hyperplane are called as the support vector points and the distance of the vectors from the hyperplane are called the margins. The basic intuition to develop over here is that more the farther SV points, from the hyperplane, more is the probability of correctly classifying the points in their respective region or classes. SV points are very critical in determining the hyperplane because if the position of the vectors changes the hyperplane’s position is altered. 


# Support Vector Classifier
We are given a training dataset of $n$ points, each point belongs to one from two classes. We want to find the "maximum-margin hyperplane" that divides all points into two groups.

![tekst alternatywny](https://drive.google.com/uc?export=view&id=1TeUlkgWC6jF3CQsxmz0Hu2ZJE7352WNk)



## Hard-margin
If the training data is linearly separable, we can select two parallel hyperplanes that separate the two classes of data, so that the distance between them is as large as possible. The region bounded by these two hyperplanes is called the "margin", and the maximum-margin hyperplane is the hyperplane that lies halfway between them. Two parallel hyperplanes are dependant on two points from other classes that are closest to each other. This might sometimes cause a situation presented on an image below (only one point from one class is close to points from the other class).

![tekst alternatywny](https://drive.google.com/uc?export=view&id=1JBI587Trkxwig2-bi8nFRUiNgsx2H0Ys)


Here the red decision boundary perfectly separates all the training points. However, is it really a good idea of having a decision boundary with such less margin? Do you think such kind of decision boundary will generalize well on unseen data? The answer is: No. The green decision boundary has a wider margin that would allow it to generalize well on unseen data. In that sense, soft margin formulation would also help in avoiding the overfitting problem.

##Soft-margin
This idea is based on a simple premise: allow SVM to make a certain number of mistakes and keep margin as wide as possible so that other points can still be classified correctly. This can be done simply by modifying the objective of SVM. Sometimes it can be helpful to allow for errors in the training set, because it may produce a more generalizable model when applied to new datasets.
In this new setting, we would aim to minimize the following objective:

$$
L = \frac{1}{2} ||w||^2 + C(number~of~mistakes)
$$

where $c$ is a parameter that decides the trade-off between maximizing the margin and minimizing the mistakes. When C is small, classification mistakes are given less importance and focus is more on maximizing the margin, whereas when C is large, the focus is more on avoiding misclassification at the expense of keeping the margin small.

#Support Vector Machine

##Kernel functions
Kernel functions are generalized functions that take two vectors (of any dimension) as input and output a score that denotes how similar the input vectors are. A simple Kernel function you already know is the dot product function: if the dot product is small, we conclude that vectors are different and if the dot product is large, we conclude that vectors are more similar.



##Main idea
Support Vector Machines use Kernel Functions to systematically find Support Vector Classifiers in higher dimensions. How?
Whereas the original problem may be stated in a finite-dimensional space, it often happens that the sets to discriminate are not linearly separable in that space. For this reason, it was proposed that the original finite-dimensional space be mapped into a much higher-dimensional space, presumably making the separation easier in that space. To keep the computational load reasonable, the mappings used by SVM schemes are designed to ensure that dot products of pairs of input data vectors may be computed easily in terms of the variables in the original space, by defining them in terms of a kernel function $k(x,y)$ selected to suit the problem. The hyperplanes in the higher-dimensional space are defined as the set of points whose dot product with a vector in that space is constant, where such a set of vectors is an orthogonal (and thus minimal) set of vectors that defines a hyperplane. Kernel functions only calculate the relationships between every pair of points as if they are in the higher dimensions; they don't actually do the transformation. It reduces the amount of computation required for SVM by avoiding the math that transforms the data from low to high dimensions.

##Summary

*  Start with data in a relatively low dimension
*  Move the data into a higher dimension
*  Find a Support Vector Classifier that separates the higher dimensional data into two groups

# QSVM

The biggest difference in QSVM compared to classical SVM is that we calculate Kernel functions by using quantum algorithms. The rest of the mechanism basicly stays the same (and it is classical aswell).

First we have to prepare the initial state, that is translating classical data points into quantum datapoints $|\phi(\overrightarrow x)\rangle$. To achieve this we have to apply a quantum circuit to the function $\phi$ and we get $V(\phi(x))$. Then we can make a measurment which returns  classical values 0 or 1 for each input datapoint $\overrightarrow x$

Following these steps we can define an ansatz for this kind of problem which is $V(\phi (\overrightarrow x))|0\rangle$. These kind of ansätze are called quantum variational circuits.

* $\phi()$ - classical function that defines feature mapping (function that get our data to higher dimension)

* $V(\phi(x))$ - circuit that translates classical datapoint to quantum datapoint. It entangles qubits and applies $\phi()$ function on them. We use ZZFeatureMap circuit with full entanglement and one repetition in our project.

*  Kernel $K(\overrightarrow x,\overrightarrow z)$ - Kernel in QSVM works like kernel in classical SVM. It's dot product/ projection of a state $\left|\phi(\overrightarrow z)\right\rangle$ to a state $\left|\phi(\overrightarrow x)\right\rangle$ <b>(Projection of states that went through feature map)</b>. That means:

<center>$K(\overrightarrow x,\overrightarrow z) = |\langle\phi(\overrightarrow x)|\phi(\overrightarrow z)\rangle|^2$ </center>

## Quantum feature maps

Let's define two auxiliary classical functions:
* $\phi_u(\overrightarrow x)=x_u$
*$\phi_{uv}(\overrightarrow x)= (\pi-x_u)(\pi-x_v)$

Then we'll use these functions in our quantum model. We can understand qubit set on quantum computer as system of interacting spin particles with spin $\frac{1}{2}$. Generally, systems like this are represented as graph e.g.
![complicated graph image](https://drive.google.com/uc?export=view&id=1EKE7MIngy9C7UMiofmBqXLTkuiHqy0z3)
Nodes $x_i$ of this graph correspond to spin particles. Edges are telling us about interactions between particles. And for such graphs nearest neighbors impact models are considered (nodes connected directly).

Qubits in our quantum computer can be presented as a simple graph:

![simple graph image](https://drive.google.com/uc?export=view&id=1TXSnML7TfKBbA9_7w-J9xM6qsPBuJkf6)



For this kind of line up we can use one dimensional Ising model, where we make use of qubits' spins and their interaction between nearest neighbours:
$x_1 - x_2, x_2 - x_3, x_3 - x_4$ and so on...
For example, to make use of two qubits we need to take to account their individual states $x_1, x_2$ and their interaction $x_1 - x_2$. For every element we have to apply Pauli Z matrix and we are ready to note the necessary hamiltonian:

$H = x_1Z_1 + x_2Z_2 + x_1x_2Z_1Z_2$

And for n qubits: 

$H_n = \Sigma_nx_nZ_n + \Sigma x_ux_vZ_uZ_v$

where $u$ and $v$ mean nearest neighbours. Now we can make use of mentioned earlier function $\phi (x)$. Our hamiltonian then looks as follows:

$H_n(\phi(\overrightarrow x)) = \Sigma_nx_nZ_n + \Sigma (\pi  - x_u)(\pi - x_v)Z_uZ_v$

Hamiltonian like this one allows us to create a unitary operation (which can be used in quantum computing) by applying it into a complex exponent as follows:

$U(\phi(\overrightarrow x)) = exp(i(\Sigma_nx_nZ_n + \Sigma (\pi  - x_u)(\pi - x_v)Z_uZ_v))$

And we have created a quantum feature map. For our purpose we have to apply first a Hadamard gate to every initial qubit and repeat whole procedure one more time. Now we can write down equation for our quantum circuit $V$.

$V(\phi(\overrightarrow x)) = U(\phi(\overrightarrow x))H^n  U(\phi(\overrightarrow x))H^n$

where $H^n$ means applying a Hadamard gate for each qubit. Now we can right down equation for qubit state affected by the circuit.

$|\phi(\overrightarrow x)\rangle = V(\phi(\overrightarrow x))|0\rangle$

By applying a hermitian conjugate on $|\phi(\overrightarrow x)\rangle$ we can get an inner product of two qubits which is required to calculate Kelner function.

$K(\overrightarrow x,\overrightarrow z) = |\langle\phi(\overrightarrow x)|\phi(\overrightarrow z)\rangle|^2$

And a graph for our circuit looks as follows:

![circuit](https://miro.medium.com/max/700/1*oYxRrkgiO4qYrAiGYroo-A.png)

By repeating this algorithm many times and using statistics we can get a square of aplitude of Kernel function we want by measuring the result. The value of Kernel function is then used by classical Support Vector Macgines.

##Making input

```
training_size=20
test_size=34
feature_dim=3
sample_Total, training_input, test_input, class_labels = breast_cancer(
    training_size=training_size,
    test_size=test_size,
    n=feature_dim,
    plot_data=True
)

X_train = np.concatenate((training_input['A'], training_input['B']), axis=0)
X_test = np.concatenate((test_input['A'], test_input['B']), axis=0)

y_train = np.concatenate((np.ones(training_size), np.zeros(training_size)))
y_test = np.concatenate((np.ones(test_size), np.zeros(test_size)))


X_train, y_train = shuffle(X_train, y_train, random_state=0)
X_test, y_test = shuffle(X_test, y_test, random_state=0)
```

##Classical models


```
# Logistic Regression
LR = LogisticRegression()
LR.fit(X_train, y_train)
LR_accuracy = LR.score(X_test, y_test)


# Decision Tree
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
dtree_accuracy = decision_tree.score(X_test, y_test)


# Support Vector Machine
SVM = SVC(probability = True)
SVM.fit(X_train, y_train)
SVM_accuracy = SVM.score(X_test, y_test)


# Linear Discriminant Analysis
LDA = LinearDiscriminantAnalysis()
LDA.fit(X_train, y_train)
LDA_accuracy = LDA.score(X_test, y_test)


# Quadratic Discriminant Analysis
QDA = QuadraticDiscriminantAnalysis()
QDA.fit(X_train, y_train)
QDA_accuracy = QDA.score(X_test, y_test)


# Random Forest Classifier
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
forest_accuracy = random_forest.score(X_test, y_test)


# K-Nearest Neighbors
KNN = KNeighborsClassifier()
KNN.fit(X_train, y_train)
KNN_accuracy = KNN.score(X_test, y_test)


# Naive Bayes
bayes = GaussianNB()
bayes.fit(X_train, y_train)
bayes_accuracy = bayes.score(X_test, y_test)
```

##QSVM

```
# Create model and make predictions
seed = 503

feature_map = ZZFeatureMap(feature_dimension=feature_dim, reps=2, entanglement='full')
qsvm = QSVM(feature_map, training_input, test_input)

backend = BasicAer.get_backend('statevector_simulator')
quantum_instance = QuantumInstance(backend, shots=2048, seed_simulator=seed, seed_transpiler=seed)

result = qsvm.run(quantum_instance)
QSVM_accuracy = result['testing_accuracy']

print("testing success ratio: {}".format(QSVM_accuracy))

print("kernel matrix during the training:")
kernel_matrix = result['kernel_matrix_training']
img = plt.imshow(np.asmatrix(kernel_matrix),interpolation='nearest',origin='upper',cmap='bone_r')
plt.show()

feature_map.draw()
```
